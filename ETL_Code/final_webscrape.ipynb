```python
# tokenize the clean comments
import nltk
from nltk import FreqDist
from nltk import corpus
#from nltk.corpora import stopwords


#default_stopwords = set(nltk.corpus.stopwords.words('english'))

final_tokens = nltk.word_tokenize(final_clean)
comments = final_tokens
final_tokens
```

```python
# create a dictionary of the token values
#comments = range(10)
from nltk.probability import ConditionalFreqDist
from nltk.tokenize import word_tokenize
sent = final_clean
cfdist = ConditionalFreqDist()
for word in word_tokenize(sent):
    condition = len(word)
    cfdist[condition][word] += 1
    cfdist
```

```python
# Calculate frequency distribution to determine unique text indicating what the conversation topics
# not about men were revealed to be
final_fdist = nltk.FreqDist(final_tokens)

# Output top 200 words

for token, frequency in final_fdist.most_common():
    print(u'{};{}'.format(final_tokens, frequency))
```
