```python
# Scrape comments from Alita: Battle Angel, which had 7 comments.
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url = 'https://bechdeltest.com/comments/8642'
response = requests.get(url)
```

```python
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup.findAll('div')
```

```python
results = soup.findAll('div')
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('div', class_="text").text
        # Identify and return price of listing
        comment = result.div.text
        # Identify and return link to listing
        # link = result.a['href']

        # Print results only if title, price, and link are available
        if (title and comment):
            print('-------------')
            print(title)
            print(comment)
    except AttributeError as e:
        print(e)
```

```python
# form a wordcloud based on the comments
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# select comments
text = ("The movie passes based on Avery brief scene showing the first meeting between Alita and Dr Chiren Chiren comments on her arm and Alita rebuffs here It’s basically limited to two lines of dialogue but it does pass In the very beginning when alita wakes up the nurse I forget her name but she is named says hello sleepyhead and alita responds hi The scene is significant because Chiren is irritated that her ex kept their daughter's prosthetic body and gave it to some girl Alita has a training sequence with Gelda her female mentor that is about technique and tenacity")
```

```python
# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)
```

```python
# Display the generated image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
